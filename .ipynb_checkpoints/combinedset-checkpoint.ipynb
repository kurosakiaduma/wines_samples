```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df_red = pd.read_csv('./winequality-red.csv')
df_white = pd.read_csv('./winequality-white.csv')                                        
```

```python
df_red.head()
```

```python
df_white.head()
```

```python
df_wines = pd.concat([df_red, df_white], ignore_index=True)
```

```python
df_wines = df_wines.rename({'wine_color': 'wine color'}, axis="columns")
df_wines
```

```python
df_wines.hist(figsize=(10,10))
```

```python
df_wines.info()
```

```python
df_wines.plot('volatile acidity', 'quality', kind='scatter', figsize=(10,10))
```

```python
df_wines.plot('residual sugar', 'quality', kind='scatter', figsize=(10,10))
```

```python
df_wines.plot('pH', 'quality', kind='scatter', figsize=(10,10))
```

```python
df_wines.plot('alcohol', 'quality', kind='scatter', figsize=(10,10), title='Scatter plot of alcohol v quality')
```

```python
df_wines.groupby(['wine color', 'pH'], as_index=False)['quality'].mean()
```

```python
df_wines.groupby('wine color', as_index=False)['quality'].mean()
```

```python
pd.cut(np.array([1,3,4,5]),4)
```

```python
pHrating = pd.qcut(df_wines['pH'], q=4, labels=['High', 'Moderate', 'Medium', 'Low'])
```

```python
df_wines['pH rating'] = pHrating
```

```python
df_wines
```

```python
df_wines.to_csv('./winequality-edited.csv', index=False)
```

```python
df_wines.groupby('pH rating', as_index=False)['quality'].mean()
```

```python
alc_median = df_wines['alcohol'].median()
df_low = df_wines.query('alcohol < @alc_median')
df_high = df_wines.query('alcohol >= @alc_median')
df_low['quality'].mean()
```

```python
df_high['quality'].mean()
```

```python
swt_median = df_wines['residual sugar'].median()
df_sweet = df_wines.query('`residual sugar` >= @swt_median')
df_bitter = df_wines.query('`residual sugar` < @swt_median')
df_bitter['quality'].mean()
```

```python
df_sweet['quality'].mean()
```

```python
colors = ['red', 'white']
df_wines.groupby('wine color')['quality'].mean().plot(kind='bar', title='Average Wine Quality by Color', color=colors, alpha=.9)
plt.xlabel('Colors', fontsize=18)
plt.ylabel('Quality Rating', fontsize=18)
```

```python
df_wines['rank_wines'] = df_wines['quality'].rank()
```
